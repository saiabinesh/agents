```python
# First let's create some things:

fruits = ["Apples", "Bananas", "Pears"]


book1 = {"title": "Great Expectations", "author": "Charles Dickens"}
book2 = {"title": "Bleak House", "author": "Charles Dickens"}
book3 = {"title": "An Book By No Author"}
book4 = {"title": "Moby Dick", "author": "Herman Melville"}

books = [book1, book2, book3, book4]
```

# Part 1: List and dict comprehensions

```python
# Simple enough to start

for fruit in fruits:
    print(fruit)
```

```python
# Let's make a new version of fruits

fruits_shouted = []
for fruit in fruits:
    fruits_shouted.append(fruit.upper())

fruits_shouted
```

```python
# You probably already know this
# There's a nice Python construct called "list comprehension" that does this:

fruits_shouted2 = [fruit.upper() for fruit in fruits]
fruits_shouted2
```

```python
# But you may not know that you can do this to create dictionaries, too:

fruit_mapping = {fruit: fruit.upper() for fruit in fruits}
fruit_mapping
```

```python
# you can also use the if statement to filter the results

fruits_with_longer_names_shouted = [fruit.upper() for fruit in fruits if len(fruit)>5]
fruits_with_longer_names_shouted
```

```python
fruit_mapping_unless_starts_with_a = {fruit: fruit.upper() for fruit in fruits if not fruit.startswith('A')}
fruit_mapping_unless_starts_with_a
```

```python
# Another comprehension

[book['title'] for book in books]
```

```python
# This code will fail with an error because one of our books doesn't have an author
print(book1.get('author'))
```

```python
# But this will work, because get() returns None

[book.get('author') for book in books]
```

```python
# And this variation will filter out the None

[book.get('author') for book in books if book.get('author')]
```

```python
# And this version will convert it into a set, removing duplicates

set([book.get('author') for book in books if book.get('author')])
```

```python
# And finally, this version is even nicer
# curly braces creates a set, so this is a set comprehension

{book.get('author') for book in books if book.get('author')}
```

```python
# First define a generator; it looks like a function, but it has yield instead of return

import time

def come_up_with_fruit_names():
    for fruit in fruits:
        time.sleep(1) # thinking of a fruit
        yield fruit
```

```python
# Then use it

for fruit in come_up_with_fruit_names():
    print(fruit)
```

```python
# Here's another one

def authors_generator():
    for book in books:
        if book.get("author"):
            yield book.get("author")
```

```python
# Use it

for author in authors_generator():
    print(author)
```

```python
# Here's the same thing written with list comprehension

def authors_generator():
    for author in [book.get("author") for book in books if book.get("author")]:
        yield author
```

```python
# Use it

for author in authors_generator():
    print(author)
```

```python
# Here's a nice shortcut
# You can use "yield from" to yield each item of an iterable

def authors_generator():
    yield from [book.get("author") for book in books if book.get("author")]
```

```python
# Use it

# for author in authors_generator():
#     print(author)
print(next(authors_generator()))
```

## Key Concept: Generator Objects

**Important**: When you call a generator function, it returns a **generator object**. Each time you call the function, you get a **new, independent** generator object.

- `authors_generator()` → Creates generator #1 (starts from beginning)
- `authors_generator()` → Creates generator #2 (starts from beginning) 
- `authors_generator()` → Creates generator #3 (starts from beginning)

To get the next item from the same generator, you must:
1. Store the generator in a variable: `gen = authors_generator()`
2. Call `next()` on that same variable: `next(gen)`

The generator object **remembers its position**, so each call to `next(gen)` gives you the next item in the sequence.


```python
# And finally - we can replace the list comprehension with a set comprehension

def unique_authors_generator():
    yield from {book.get("author") for book in books if book.get("author")}
```

```python
# Use it

for author in unique_authors_generator():
    print(author)
```

```python
# And for some fun - press the stop button in the toolbar when bored!
# It's like we've made our own Large Language Model... although not particularly large..
# See if you understand why it prints a letter at a time, instead of a word at a time. If you're unsure, try removing the keyword "from" everywhere in the code.

import random
import time

pronouns = ["I", "You", "We", "They"]
verbs = ["eat", "detest", "bathe in", "deny the existence of", "resent", "pontificate about", "juggle", "impersonate", "worship", "misplace", "conspire with", "philosophize about", "tap dance on", "dramatically renounce", "secretly collect"]
adjectives = ["turqoise", "smelly", "arrogant", "festering", "pleasing", "whimsical", "disheveled", "pretentious", "wobbly", "melodramatic", "pompous", "fluorescent", "bewildered", "suspicious", "overripe"]
nouns = ["turnips", "rodents", "eels", "walruses", "kumquats", "monocles", "spreadsheets", "bagpipes", "wombats", "accordions", "mustaches", "calculators", "jellyfish", "thermostats"]

def infinite_random_sentences():
    while True:
        yield from random.choice(pronouns)
        yield " "
        yield from random.choice(verbs)
        yield " "
        yield from random.choice(adjectives)
        yield " "
        yield from random.choice(nouns)
        yield ". "

for letter in infinite_random_sentences():
    print(letter, end="", flush=True)
    time.sleep(0.02)
```

## Quick Refresher: Python Classes and Methods

Before the exercise, here's a quick refresher on classes:


```python
# 2. Methods that return True/False (boolean methods)
# These typically start with 'is', 'has', 'can', etc.

class Animal:
    def __init__(self, name, species, tail_length=None):
        self.name = name
        self.species = species
        self.tail_length = tail_length  # Can be None if no tail
    
    # Boolean method - returns True or False
    def has_tail(self):
        # Check if tail_length exists (is not None)
        return self.tail_length is not None
    
    # Another boolean method
    def is_mammal(self):
        return self.species.lower() in ['dog', 'cat', 'horse', 'human']

animal1 = Animal("Fluffy", "dog", tail_length=5)
animal2 = Animal("Whiskers", "cat", tail_length=None)

print(animal1.has_tail())  # Returns True
print(animal2.has_tail())  # Returns False
print(animal1.is_mammal())  # Returns True
```

```python
# 3. Classes that contain lists and generator methods

class Garage:
    def __init__(self, name):
        self.name = name
        self.cars = []  # List to store car instances
    
    # Method to add items to the list
    def add_car(self, car):
        self.cars.append(car)
    
    # Generator method - uses 'yield' instead of 'return'
    def get_cars(self):
        # Yield each car one at a time
        for car in self.cars:
            yield car
    
    # Another generator method with filtering
    def get_old_cars(self):
        # Only yield cars that are old
        for car in self.cars:
            if car.is_old():
                yield car

# Create a garage and add cars
garage = Garage("My Garage")
garage.add_car(Car("Toyota", "Corolla", 2020))
garage.add_car(Car("Ford", "Mustang", 2005))
garage.add_car(Car("Honda", "Civic", 2008))

# Use the generator method
print("All cars:")
for car in garage.get_cars():
    print(car.get_info())

print("\nOld cars only:")
for car in garage.get_old_cars():
    print(car.get_info())
```

```python
# 4. Generator method that returns unique values
# Use a set to track what you've already yielded

class School:
    def __init__(self, name):
        self.name = name
        self.students = []
    
    def add_student(self, student):
        self.students.append(student)
    
    # Generator that yields unique values
    def unique_teachers(self):
        seen = set()  # Track what we've already yielded
        for student in self.students:
            teacher = student.teacher
            if teacher and teacher not in seen:
                seen.add(teacher)
                yield teacher

# Simple Student class for the example
class Student:
    def __init__(self, name, teacher):
        self.name = name
        self.teacher = teacher

# Create school and add students
school = School("Python High")
school.add_student(Student("Alice", "Mr. Smith"))
school.add_student(Student("Bob", "Mr. Smith"))  # Same teacher
school.add_student(Student("Charlie", "Ms. Jones"))
school.add_student(Student("Diana", "Mr. Smith"))  # Same teacher again

# Get unique teachers
print("Unique teachers:")
for teacher in school.unique_teachers():
    print(teacher)
# Output: Mr. Smith, Ms. Jones (only unique ones!)
```

```python
# 5. Optional parameters and handling None values

class Recipe:
    def __init__(self, name, ingredients, chef=None):
        # Parameters can have default values (like chef=None)
        self.name = name
        self.ingredients = ingredients
        self.chef = chef  # Can be None if no chef specified
    
    # Method to check if something exists (not None)
    def has_chef(self):
        # Return True if chef is not None, False otherwise
        return self.chef is not None
    
    def get_info(self):
        info = f"Recipe: {self.name}"
        if self.has_chef():
            info += f" by {self.chef}"
        return info

# Create recipes with and without chefs
recipe1 = Recipe("Pasta", ["noodles", "sauce"], chef="Mario")
recipe2 = Recipe("Soup", ["water", "vegetables"])  # No chef (None)

print(recipe1.has_chef())  # True
print(recipe2.has_chef())  # False
print(recipe1.get_info())
print(recipe2.get_info())
```

## Key Points Summary:

1. **Class Definition**: `class MyClass:`
2. **Constructor**: `def __init__(self, param1, param2=None):` - called when creating instances
3. **Instance Variables**: `self.variable_name = value` - store data in the object
4. **Methods**: `def method_name(self):` - functions that belong to the class
5. **Boolean Methods**: Return `True` or `False`, often check if something exists: `return self.value is not None`
6. **Generator Methods**: Use `yield` instead of `return` to yield items one at a time
7. **Unique Values**: Use a `set()` to track what you've already seen/yielded
8. **Optional Parameters**: Use default values like `param=None` in `__init__`
9. **Lists in Classes**: Initialize with `self.items = []` in `__init__`
